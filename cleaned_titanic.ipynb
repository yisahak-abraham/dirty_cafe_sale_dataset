```python
from google.colab import drive
drive.mount('/content/drive')
import os
import pandas as pd
import numpy as np

titanic_path = '/content/drive/MyDrive/Titanic-Dataset.csv'
os.path.isfile(titanic_path)
train_df = pd.read_csv(titanic_path)
print("The information of data of first columns")
print("="*50)
print(train_df.head(10))
print("="*50)
print("The info data")
train_df.info()
print("The statistcial of data ")
print(train_df.describe())
print("The missing vaues of data ")
print(train_df.isnull().sum())
age_median = train_df["Age"].median()
print(f"Median Age to be used: {age_median}")
train_df["Age"] = train_df["Age"].fillna(age_median)
print(f"Missing age values: {train_df["Age"].isnull().sum()}")
train_df.select_dtypes(include=["number"]).describe()
print(train_df[["Age", "Fare", "SibSp", "Parch"]].describe())
train_df["Pclass"].value_counts()
train_df["Sex"].value_counts()
train_df["Cabin"] = train_df["Cabin"].fillna("U")
print(train_df["Cabin"])
train_df["Sex"] = train_df["Sex"].map({"female": 1, "male": 0})
print(train_df["Age"].isnull().sum())
train_df["Embarked"].value_counts()
train_df['Embarked'].isnull().sum()
train_df.to_csv("cleaned_titanic.csv", index=False)
from google.colab import files
files.download("cleaned_titanic.csv")
train_df.to_csv("/content/drive/MyDrive/cleaned_titanic.csv", index=False)


```
